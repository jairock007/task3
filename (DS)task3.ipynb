```python
print(df['cylindernumber'].unique()) 
```

```python
df['cylindernumber'] = df['cylindernumber'].replace({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8})
```

```python
print(df['cylindernumber'].unique())
```

```python
print(df['doornumber'].unique()) # We can't work on categorical data( categories or groups or NO numbers) 
```

```python
df['doornumber'] = df['doornumber'].replace({'four':4, 'two':2})
```

```python
print(df['doornumber'].unique()) # we only work on Quantitative data(numbers)
```

```python
df = df.drop(['car_ID','CarName'], axis=1) # people don't buy car because of CarName and Car_ID
```

```python
df.head()
```

```python
y = df['price']
x = df.drop('price',axis=1)
```

```python
#Encoding all categorical data
x = pd.get_dummies(x)
x.head()
```

```python
ss = StandardScaler()

x = pd.DataFrame(ss.fit_transform(x), columns=x.columns)
x.head()
```

```python


x_tr, x_test, y_tr, y_test = train_test_split(x,y, test_size=0.3, random_state=47) # for scikit linear regression

```

```python


lr = LinearRegression()
lr.fit(x_tr, y_tr)

```

```python


y_predict_test_lr = lr.predict(x_test)
y_predict_train_lr = lr.predict(x_tr)

```

```python


plt.plot(y_test, y_predict_test_lr, 'r.')
plt.plot(y_tr, y_predict_train_lr, 'bx')
plt.legend(['Prediction on Test Data', 'Prediction on Training Data'])
plt.show()

```

```python


print("Score on test data vs training data")
metrics.r2_score(y_test, y_predict_test_lr)*100,metrics.r2_score(y_tr, y_predict_train_lr)*100

```
